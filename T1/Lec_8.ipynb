```python
import pandas as pd
import plotly.express as px

data = {
    "Month": ["Jan","Feb","Mar","Apr","Jan","Feb","Mar","Apr"],
    "Sales": [100,120,140,160,90,110,130,150],
    "Region": ["North","North","North","North",
               "South","South","South","South"]
}

df = pd.DataFrame(data)
fig = px.line(
    df,
    x="Month",
    y="Sales",
    color="Region",
    markers=True,
   width=800,
    height=500,
    title="Monthly Sales Comparison",
    color_discrete_sequence=["blue",.
                             "red"]
)

fig.show()
```

```python
data = {
    "Marks":[40,45,50,55,60,65,70,75,80,85],
    "Gender":["Male","Female","Male","Female","Male",
              "Female","Male","Female","Male","Female"]
}

df = pd.DataFrame(data)
fig = px.histogram(
    df,
    x="Gender",
    y='Marks',
    nbins=5,
    color="Gender",
    histfunc="sum",
    pattern_shape="Gender",
    text_auto=True,
    color_discrete_map={"Male":"blue","Female":"pink"},
#     color_discrete_sequence=["blue","pink"] #Optional
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px

data = {
    "Department": ["IT", "HR", "Finance", "Marketing"],
    "Employees": [40, 20, 25, 15]
}

df = pd.DataFrame(data)
fig = px.pie(
    df,
    values="Employees",
    names="Department",
    color_discrete_sequence=["blue", "green", "orange", "red"],
    title="Employee Distribution by Department"
)

fig.update_traces(
    textinfo="percent+label",
    textposition="inside",# auto ,pull,inside
    pull=[0,0.1,0,0]
)
fig.show()
```
