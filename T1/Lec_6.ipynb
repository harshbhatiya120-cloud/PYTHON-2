```python
def grade(total):
    if total>=250:
        return 'A'
    elif total>=200:
        return "B"
    else:
        return "C"
df1['Grade']=df1['total'].apply(grade)
df1        
```

```python
df2=pd.DataFrame({'width':[1,2],'height':[3,4]})
def get_stats(row):
    return [row['width']*row['height'],2*(row['width']+row['height'])]
df2
```

```python
df2.apply(get_stats,axis=1)
```

```python
df2.apply(get_stats,axis=1,result_type='expand')
```

```python
df2.apply(get_stats,axis=1,result_type='reduce')
```

```python
df2.apply(get_stats,axis=1,result_type='broadcast')
```

```python
df1=pd.DataFrame({
    'employee':['a','b','c'],
    'department':['it','hr','finance']
})
df2=pd.DataFrame({
    'employee':['a','b','d'],
    'salary':[50000,40000,30000]

})
r1=pd.merge(df1,df2,on='employee',how='inner')
print(r1)
```

```python
r1=pd.merge(df1,df2,on='employee',how='outer')
print(r1)
```

```python
r1=pd.merge(df1,df2,on='employee',how='left')
print(r1)
```

```python
r1=pd.merge(df1,df2,on='employee',how='right')
print(r1)
```
